# EDA


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='N_Days', hue='Status', multiple="stack", bins=20,
             kde=True, palette={'C': 'magenta', 'CL': 'yellow', 'D': 'red'})

plt.xlabel('Continuous Data')
plt.ylabel('Frequency')
plt.title('Histogram of Continuous Data by Category')

plt.legend(title='Status')

plt.show()
```

from 0 to 1000 , it's higher probability to be D
otherwise the highest prob is C , CL appears on the interval 500 to 2700
